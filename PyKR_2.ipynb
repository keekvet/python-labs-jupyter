# Злочиність у США з 1975 по 2015 (про Україну не знайшов)

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
table = pd.read_csv('dataset.csv')
table
```

# Групую за роками щоб відкинути інформацію про окремі штати та міста

```python
per_year = table.groupby('report_year')

crimes_percapita = per_year.mean()[
    ['crimes_percapita', 
     'homicides_percapita', 
     'rapes_percapita', 
     'assaults_percapita', 
     'robberies_percapita']]

per_year = per_year.sum()

per_year = crimes_percapita.join(per_year[per_year.columns.difference(crimes_percapita.columns)])


per_year
```

# Співідношення усіх злочинів та тяжких злочинів


```python

fig, ax1 = plt.subplots()

ax1.set_xlabel('рік')
ax1.set_ylabel('відносна кількість злочинів')
ax1.plot(per_year.index, per_year['crimes_percapita'], c = 'r')
ax1.plot(per_year.index, per_year['crimes_percapita'], c = 'g')

ax2 = ax1.twinx()
ax2.set_ylabel('кількість тяжких злочинів')
ax2.plot(per_year.index, per_year['violent_crimes'], c = 'r')

plt.grid()
ax1.legend(['тяжкі злочини', 'загальне співвідношення'])

plt.title('щорічна кількість злочинів у США')

```

# Ріст населення та злочини
на графіку видно що кількість населення та кілкість злочинів не пов'язані

```python

fig, ax1 = plt.subplots()

ax1.set_xlabel('рік')
ax1.set_ylabel('відносна кількість злочинів')
ax1.plot(per_year.index, per_year['crimes_percapita'], c = 'r', linestyle = ':')

ax2 = ax1.twinx()
ax2.set_ylabel('загальне населення')
ax2.plot(per_year.index, per_year['population'], c = 'b')

plt.grid()
plt.title('кількість злочинів у США та населення')
```

# Відношення видів злочинів від 1975 до 2015

```python
types_names = [
    'assaults_percapita',
    'robberies_percapita',
    'rapes_percapita',
    'homicides_percapita'
]

crimes_types =  per_year[types_names]

types_names_edited = [s.replace('_', ' ') for s in types_names]

crimes_types = crimes_types.mean(axis = 0)
crimes_types

plt.pie(crimes_types, autopct='%1.1f%%')
plt.legend(types_names_edited, title='види злочинів', bbox_to_anchor=(1.05, 1), loc='upper left')
```

# Види злочинів по роках

```python
colors = ['r', 'gold', 'steelblue', 'black']

for i in range(0, len(types_names)):
    plt.bar(per_year.index, per_year[types_names[i]], color=colors[i])
    
plt.legend(types_names_edited, title='види злочинів', bbox_to_anchor=(1.05, 1), loc='upper left')
```

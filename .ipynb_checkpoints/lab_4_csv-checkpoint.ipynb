```python
import pandas as pd

frame = pd.read_csv('olympics.csv', index_col = 0, skiprows = 1)
frame
```

```python
frame = frame.rename(columns = {'01 !' : 'Gold', '02 !' : 'Silver', '03 !' : 'Bronze'})
```

```python
frame[['Gold', 'Silver', 'Bronze']]
```

```python
frame = frame.drop('Totals')
```

```python
frame[frame['Gold'] > 0]
```

```python
frame[frame['Gold'] > 0]['Gold'].count()
```

```python
frame.where(frame['Gold'] == frame['Gold'].max()).dropna()
```

```python
frame.where(frame['Gold'] > 0).where(frame['Silver'] > 0).dropna()
```

```python
frame['country'] = frame.index
frame.index = frame['Gold']
```

```python
frame[frame['country'] == 'Ukraine\xa0(UKR)']
```

```python
frame['sum summer'] = frame[['Gold', 'Silver', 'Bronze']].sum(axis=1)
frame
```

```python
frame = frame.sort_values('sum summer', ascending = False)
frame
```

```python
frame = frame.head(7)
```

```python
frame
```

```python
import matplotlib.pyplot as plt
medals_summer_pie = plt.pie(frame['sum summer'],  autopct = '%.1f%%', labels = frame['country'])
```

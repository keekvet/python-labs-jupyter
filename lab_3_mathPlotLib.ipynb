```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
f_sin = [math.sin(i / 10) for i in range(0, 100)]
f_cos = [math.cos(i / 10) for i in range(0, 100)]

plt.grid()

plt.title('sin and cos')

plt.xlabel('xxxxxx')
plt.ylabel('yyyyyy')

plt.plot(f_sin)
plt.plot(f_cos)

plt.grid(axis = 'x', color = 'y', alpha = 0.9, linewidth = 5)

plt.legend(['sin', 'cos'])
```

```python
plt.plot(f_sin, '|')
```

```python
plt.plot(f_sin, lw = 10)
```

```python
plt.fill_between(range(0, 100), f_sin)
```

```python
plt.scatter(f_sin, f_cos, marker = '*', s = 100)
```

```python
func_up = [i for i in range(0, 10)] * 2
func_down = [ 10 - i for i in range(0, 10)] * 2
plt.bar(range(0, 20),  func_up)
```

```python
plt.bar([i - 0.2 for i in range(0, 20)], func_down, width = 0.4)
plt.bar([i + 0.2 for i in range(0, 20)], func_up, width = 0.4)
```
